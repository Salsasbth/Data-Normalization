# <p style="background-color:#FFC0CB;font-family:newtimeroman;color:#222021;font-size:130%;text-align:center;border-radius:10px 10px;margin-bottom:10px;">**4. Data Cleaning**</p>

#### a. Manipulasi kolom meliputi penggantian nama kolom, penghapusan kolom, atau penambahan kolom

##### - Menghapus kolom 'Month Number', 'Month Name', dan 'Year'

````python
df.drop(columns = df.columns[13:16], inplace = True)
df.head()
````

Kolom yang  tidak digunakan atau tidak berpengaruh pada analisa akan dikeluarkan atau dihapus, pada data frame ini akan dihapus kolom 'Month Number', 'Month Name', dan 'Year'

##### - Mengganti nama kolom
Mengganti beberapa kolom yang memiliki whitespace pada nama kolom dan mengganti spasi dengan '_'

````python
df = df.rename(columns={' Product ': 'Product', 
                   ' Discount Band ': 'Discount_Band',
                   ' Units Sold ': 'Units_Sold',
                   ' Manufacturing Price ': 'Manufacturing_Price',
                   ' Sale Price ': 'Sale_Price',
                   ' Gross Sales ': 'Gross_Sales',
                   ' Discounts ': 'Discounts',
                   '  Sales ': 'Sales',
                   ' COGS ': 'COGS',
                   ' Profit ': 'Profit'
                   })
df.columns
````

Menghilangkan whitespace pada nama kolom dan mengganti spasi menjadi '_'

##### - Menghilangkan simbol '$' dan '()' pada kolom :  
'Units_Sold', 'Manufacturing_Price', 'Sale_Price', 'Gross_Sales', 'Discounts', 'Sales', 'COGS', 'Profit'

````python
# Tabel tersebut berada pada index ke 4 sampai 11
df[df.columns[4:12]] = df[df.columns[4:12]].replace(to_replace = r'[ ()$,]', value = '', regex = True)
df.head()
````

Menghilangkan spasi dan simbol '$', '(', dan ')' pada value data frame agar kemudian kolom yang tipe datanya tidak sesuai dapat disesuaikan.

#### b. Check dan handling *missing value*

Pada data frame, list *missing value* tidak menggunakan yang standar, sehingga harus didefinisikan dan diubah menjadi list standar *missing value*

````python
# Mendefinisikan list non-standard missing values yang mungkin terdapat dalam kolom
non_standard_missing_values = ['NA', 'N/A', 'null', '-']

# Mengganti non-standard missing values dengan standar yaitu NaN
df.replace(non_standard_missing_values, pd.NA, inplace=True)

# Menampilkan banyak missing values pada tiap kolom
df.isnull().sum()
````

Mengganti *missing value* dengan angka 0, sehingga data dapat di proses

````python
df = df_replaced2 = df.replace(to_replace=[np.nan], value = 0)
df
````

````python
# Check missing value
df.isnull().sum()
````

#### c. Penggantian tipe data yang sesuai dengan seharusnya.

````python
# Mengganti tipe data pada kolom Manufacturing_Price, Sale_Price, Gross_Sales, Discounts, Sales, COGS, dan Profit
# (index kolom ke 5 sampai 11) menjadi float
df[df.columns[5:12]] = df[df.columns[5:12]].astype(float)

# Mengganti tipe data pada kolom Unit_Sold (index kolom ke 4) menjadi integer
df['Units_Sold'] = df['Units_Sold'].astype(float).astype(int)

# Mengganti tipe data pada kolom Date (index kolom ke 12) menjadi date time
df['Date'] = pd.to_datetime(df['Date'])


df.info()
````

Penggantian tipe data dilakukan agar data dapat diproses selanjutnya. Pada data frame ini kolom yang tipe datanya akan diganti yaitu :  
- Manufacturing_Price, Sale_Price, Gross_Sales, Discounts, Sales, COGS, dan Profit dari object menjadi float  
- Units_Sold dari object menjadi integer
- Date dari object menjadi datetime

# <p style="background-color:#FFC0CB;font-family:newtimeroman;color:#222021;font-size:130%;text-align:center;border-radius:10px 10px;margin-bottom:10px;">**5. Data Normalization**</p>


Lakukan normalisasi data yang sudah diolah sebelumnya sampai dengan 3NF.

- Membuat tabel Segment ID

````python
segmentID = [f'S{i}' for i in range(len(df['Segment'].drop_duplicates()))]
segment_table = pd.DataFrame(data = {'Seg_ID' : segmentID,
                                    'Segment' : (df['Segment'].drop_duplicates()).copy()
                                    }
                            ).reset_index(drop=True)
segment_table.to_csv('Segment.csv', index=False)
segment_table
````

- Membuat tabel Country ID

````python
countryID = [f'C{i}' for i in range(len(df['Country'].drop_duplicates()))]
country_table  = pd.DataFrame(data = {'Country_ID' : countryID,
                                      'Country' : (df['Country'].drop_duplicates()).copy()
                                    }
                                ).reset_index(drop=True)
country_table.to_csv('Country.csv', index=False)
country_table
````

- Membuat tabel Product ID

````python
productID = [f'P{i}' for i in range(len(df['Product'].drop_duplicates()))]
product_table  = pd.DataFrame(data = {'Product_ID' : productID,
                                      'Product' : (df['Product'].drop_duplicates()).copy(),
                                      'Manufacturing_Price' : (df['Manufacturing_Price']).drop_duplicates().copy()
                                    }
                                ).reset_index(drop=True)
product_table.to_csv('Product.csv', index=False)
product_table
````

- Membuat tabel discount_band ID

````python
discountID = [f'D{i}' for i in range(len(df['Discount_Band'].drop_duplicates()))]
discount_table  = pd.DataFrame(data = {'Discount_ID' : discountID,
                                      'Discount_Band' : (df['Discount_Band'].drop_duplicates()).copy()
                                    }
                                ).reset_index(drop=True)
discount_table.to_csv('Discount.csv', index=False)
discount_table
````

- Main Table

````python
# Membuat tabel utama
mainTable = df[df.columns[0:13]]

# membuat series yang menyetor 
series = [f'P{i}' for i in range(len(mainTable))]

segmentID=[]
countryID=[]
productID = []
discountID = []

for i in range(len(mainTable)):
  for j in range(len(segment_table)):
    if mainTable.iloc[i,0] == segment_table.iloc[j,1]:
      segmentID.append(segment_table.iloc[j,0])
      break
  for j in range(len(country_table)):
    if mainTable.iloc[i,1] == country_table.iloc[j,1]:
      countryID.append(country_table.iloc[j,0])
      break
  for j in range(len(product_table)):
    if mainTable.iloc[i,2] == product_table.iloc[j,1]:
      productID.append(product_table.iloc[j,0])
      break
  for j in range(len(discount_table)):
    if mainTable.iloc[i,3] == discount_table.iloc[j,1]:
      discountID.append(discount_table.iloc[j,0])
      break

# Membuat table terpisah dari database tanpa referensi ke database
mainTable = mainTable.copy()

#Drop Kolom tidak berguna
mainTable.drop(columns = 'Segment', inplace = True)
mainTable.drop(columns = 'Country', inplace = True)
mainTable.drop(columns = 'Product', inplace = True)
mainTable.drop(columns = 'Discount_Band', inplace = True)

# Memasukkan kolom
mainTable.insert(0, column = 'SegmentID', value = segmentID)
mainTable.insert(1, column = 'CountryID', value = countryID)
mainTable.insert(2, column = 'ProductID', value = productID)
mainTable.insert(3, column = 'DiscountID', value = discountID)

mainTable.to_csv('MainTable.csv', index=False)

mainTable
````
